# Elite MEV Bot v2.1 Test Notebook

This notebook tests the MEV bot functionality with Claude integration.

```python
// Test Rust kernel functionality
println!("Testing Rust kernel...");
let version = "2.1";
println!("Elite MEV Bot version: {}", version);
```

```python
# Test Python kernel functionality
import os
import subprocess

print("Testing Python kernel...")
print(f"Current directory: {os.getcwd()}")

# Check if we can access the MEV bot files
mev_files = [f for f in os.listdir('.') if 'elite_mev' in f]
print(f"Found MEV files: {mev_files}")
```

```python
# Test if we can run Cargo commands
try:
    result = subprocess.run(['cargo', '--version'], capture_output=True, text=True, cwd='.')
    print(f"Cargo version: {result.stdout.strip()}")
except Exception as e:
    print(f"Error running cargo: {e}")
```

```python
# Try to build the MEV bot
try:
    print("Attempting to build elite MEV bot...")
    result = subprocess.run(
        ['cargo', 'build', '--bin', 'elite_mev_bot_v2_1_production'], 
        capture_output=True, 
        text=True, 
        cwd='.'
    )
    
    if result.returncode == 0:
        print("✅ Build successful!")
    else:
        print(f"❌ Build failed with return code: {result.returncode}")
        print(f"STDERR: {result.stderr[:1000]}..." if len(result.stderr) > 1000 else f"STDERR: {result.stderr}")
        
except Exception as e:
    print(f"Error during build: {e}")
```
